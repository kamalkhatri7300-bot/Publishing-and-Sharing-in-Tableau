Question 1. What is the key difference between publishing a dashboard on Tableau Public versus Tableau Server/Cloud in terms of accessibility and data security?

- Answer 1.

Accessibility: Tableau Public is entirely open to the world; anyone with the link can view your work. Tableau Server/Cloud is private, requiring a login and authenticated access.

Data Security: On Tableau Public, data is not secure as anyone can often download the workbook and underlying data. Tableau Server/Cloud provides enterprise-grade security, allowing you to restrict who sees what.


Questoin 2. When publishing a Tableau dashboard, what happens to a live connection compared to an extract?
How does this affect data refresh and performance?

- Answer 2

Live Connection: Tableau queries the database in real-time. If the source data changes, the dashboard updates instantly. However, performance depends on the database speed.

Extract: A static snapshot of data stored in Tableau's memory (Hyper).

Effect: Extracts generally offer faster performance for complex dashboards, but data is only as fresh as the last scheduled refresh.


Question 3. In Tableau Server or Cloud (conceptually), which feature allows administrators to control which users can view, interact with, or edit a published dashboard?

- Answer 3

The feature is Permissions (often managed through Content Roles or User/Group Permissions). Administrators use these to define capabilities like View, Filter, Download Summary Data, or Web Edit.

Question 4. If you want users to view a dashboard but not download the underlying data, which permission setting would you adjust in Tableau Server?

- Answer 4

You would adjust the "Download Full Data" permission setting. By setting this to Deny or Unspecified for specific users/groups, they can view the visuals but cannot export the raw underlying rows.

Question 5. Name two methods by which a Tableau dashboard can be embedded into a web page or website.

- Answer 5

Embed Code: Copying the JavaScript snippet provided in the "Share" dialog of a published dashboard.

Iframe: Using an HTML <iframe> tag with the dashboard's URL.

Question 6 List two best practices you should follow when preparing a dashboard for public sharing to ensure data security and privacy.

- Answer 6

Anonymize/Aggregate Data: Ensure no Personally Identifiable Information (PII) is included; use aggregated totals rather than row-level details.

Disable Data Download: Explicitly hide the "Download" buttons and underlying data sheets in the dashboard settings before publishing.



Question 7. After publishing a dashboard, users report a “Data not found” error. What is the most common cause related to data source settings?

- Answer 7

The most common cause is that the Data Source was not published alongside the workbook, or the credentials required to access the data source were not embedded, leading to an authentication failure when the server tries to reach the database.

Question 8. When embedding a Tableau dashboard into a secure internal portal, what two additional security measures should be implemented to ensure only authorized users can access the embedded content?

- Answer 8

Trusted Authentication (Single Sign-On / SSO): Using SAML, Okta, or Kerberos so users don't have to sign in twice.

Row-Level Security (RLS): Implementing filters that restrict data visibility based on the specific user’s identity, ensuring they only see data relevant to them even within a shared dashboard.
